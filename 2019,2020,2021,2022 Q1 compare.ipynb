```python
import pandas as pd
import matplotlib.pyplot as plt
# You can configure the format of the images: ‘png’, ‘retina’, ‘jpeg’, ‘svg’, ‘pdf’.
%config InlineBackend.figure_format = 'svg'
# this statement allows the visuals to render within your Jupyter Notebook
%matplotlib inline 
import seaborn as sns
import numpy as np
```

```python
data = {'Period':['2019 Q1', '2021 Q1', '2022 Q1'],
        'Ridership':[437.715515, 137.688160, 189.837580]}
 
# Create DataFrame
df = pd.DataFrame(data)
 
# Print the output.
df
```

```python
sns.barplot(df["Period"],df["Ridership"])
plt.xlabel('Time Period',fontsize=12) 
plt.ylabel('Millions',fontsize=12)
plt.title('Total Ridership of Year (19,21,22) Q1',fontsize=14)
plt.ylim(0,500) 
plt.savefig('Total Ridership of Year (19,21,22) Q1',dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.pie([0.25,0.25,0.25,0.25])
plt.savefig('Ridership distribution',dpi=300)
plt.show()
```

```python
weekly_2022 = pd.read_pickle("2022_Q1_weekly_pattern.pkl")
weekly_2022
```

```python
weekly_2019 = pd.read_pickle("2019_Q1_weekly_pattern.pkl")
weekly_2019
```

```python
# create data
x = np.arange(7)
y1 = weekly_2019["Avg_Entries_million"]
y2 = weekly_2022["Avg_Entries_million"]

width = 0.4
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, y1, width, color='silver')
plt.bar(x+0.2, y2, width, color='cornflowerblue')


plt.xticks(x, ['Mo','Tu','We','Th','Fr','St','Sn'])
plt.xlabel("Day of Week",fontsize=12)
plt.ylabel("Millions",fontsize=12)
plt.legend(["2019 Q1", "2022 Q1"])
plt.title("Average Entries by Day of Week with 2019 as Reference",fontsize=12)
plt.ylim(0,7)
plt.savefig("Compare_Average Entries by Day of Week",dpi=300)
plt.show()
```
